```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("C://Users//user//Pictures//Saved Pictures//79065_5.jpg",0)
edges = cv2.Canny(img ,100 ,200)

plt.subplot(121)
plt.imshow(img,cmap = 'gray')
plt.title('original image')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(edges,cmap = 'gray')
plt.title('canny edge image')
plt.xticks([])
plt.yticks([])

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()

```
